```python
# import the necessary packages
import pandas as pd
```

```python
# load the  dataset

baidu = pd.read_csv('../data/baidu/baidu_2023-11-27.csv')
google = pd.read_csv('../data/google/google_2023-11-26.csv')
yahoo = pd.read_csv('../data/yahoo/yahoo_2023-11-26.csv')
duckduckgo = pd.read_csv('../data/duckduckgo/duckduckgo_2023-11-26.csv')
```

```python
# change to template 1 and template 2 in baidu 
import re
# # Define the templates and their corresponding regex patterns
# template_1 = "为什么待替换词汇这么"
# template_2 = "待替换词汇为什么这么"
# Define the templates and their corresponding regex patterns
template_1 = "为什么.+(这么)"
template_2 = "(.+)为什么这么"
# Compile the regex patterns
template_1_pattern = re.compile(template_1)
template_2_pattern = re.compile(template_2)

# Filter the data based on the templates
baidu_template_1 = baidu[baidu['query'].str.contains(template_1_pattern)]
baidu_template_2 = baidu[baidu['query'].str.contains(template_2_pattern)]
```

```python
baidu_template_1 
```

```python
# Function to check if the suggestion starts with the query
def check_suggestion_starts_with_query(row):
    query = row['query']
    suggestion = row['suggestion']
    return suggestion.startswith(query)

# Apply the function to each row in the DataFrame
baidu_template_1['suggestion_starts_with_query'] = baidu_template_1.apply(check_suggestion_starts_with_query, axis=1)
baidu_template_2['suggestion_starts_with_query'] = baidu_template_2.apply(check_suggestion_starts_with_query, axis=1)
baidu['suggestion_starts_with_query'] = baidu.apply(check_suggestion_starts_with_query, axis=1)
# remove duplicates in each groups 
```

```python
# Filter the DataFrame to keep only rows where 'suggestion_starts_with_query' is True
baidu_template_1_follow_query = baidu_template_1[baidu_template_1['suggestion_starts_with_query']]
baidu_template_2_follow_query = baidu_template_2[baidu_template_2['suggestion_starts_with_query']]
# Display the first few rows of the filtered DataFrame
```

```python
def extract_after_keyword(suggestion, keyword="这么"):
    
    pos = suggestion.find(keyword)
    if pos != -1:
        # Return the part of the string including and after the keyword
        return suggestion[pos:].strip()
    return ""  # Return empty string if the keyword is not found
```

```python
baidu_template_1_follow_query.loc[:, 'extracted_suggestion'] = baidu_template_1_follow_query['suggestion'].apply(lambda x: extract_after_keyword(x))
baidu_template_2_follow_query.loc[:, 'extracted_suggestion'] = baidu_template_2_follow_query['suggestion'].apply(lambda x: extract_after_keyword(x))
```

```python
# Assuming baidu_follow_query is a filtered DataFrame and you want to add a new column to it
baidu.loc[:, 'extracted_suggestion'] = baidu['suggestion'].apply(lambda x: extract_after_keyword(x))
```

```python
def is_suggestion_valid(query, suggestion):
    # Convert both query and suggestionc to the same case (lowercase in this case)
    return suggestion.lower().startswith(query.lower())
```

```python
# Apply the function to your DataFrames
yahoo['suggestion_follows_query'] = yahoo.apply(lambda row: is_suggestion_valid(row['query'], row['suggestion']), axis=1)
google['suggestion_follows_query'] = google.apply(lambda row: is_suggestion_valid(row['query'], row['suggestion']), axis=1)
duckduckgo['suggestion_follows_query'] = duckduckgo.apply(lambda row: is_suggestion_valid(row['query'], row['suggestion']), axis=1)
baidu['suggestion_follows_query'] = baidu.apply(lambda row: is_suggestion_valid(row['query'], row['suggestion']), axis=1)
```

```python
# Filter the DataFrames to keep only rows where 'suggestion_follows_query' is True
yahoo_follow_query = yahoo[yahoo['suggestion_follows_query']]
google_follow_query = google[google['suggestion_follows_query']]
duckduckgo_follow_query = duckduckgo[duckduckgo['suggestion_follows_query']]
baidu_follow_query = baidu[baidu['suggestion_follows_query']]
```
